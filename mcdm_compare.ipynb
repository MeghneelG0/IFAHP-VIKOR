```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, kendalltau

# Create DataFrame
df = pd.DataFrame({
    "IFAHP-VIKOR": ifahp_vikor,
    "IFAHP-TOPSIS": ifahp_topsis,
    "FAHP-VIKOR": fahp_vikor,
    "FAHP-TOPSIS": fahp_topsis,
    "AHP-VIKOR": ahp_vikor,
    "AHP-TOPSIS": ahp_topsis
})

# Spearman Correlation
spearman_matrix = df.corr(method="spearman")

# Kendall Correlation
kendall_matrix = pd.DataFrame(index=df.columns, columns=df.columns)
for i in df.columns:
    for j in df.columns:
        tau, _ = kendalltau(df[i], df[j])
        kendall_matrix.loc[i, j] = tau
kendall_matrix = kendall_matrix.astype(float)

# Plotting function
def plot_heatmap(corr_matrix, title, cmap="coolwarm"):
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap=cmap, linewidths=0.5, square=True)
    plt.title(title, fontsize=14)
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Visualize Spearman and Kendall
plot_heatmap(spearman_matrix, "Spearman Correlation Matrix")
plot_heatmap(kendall_matrix, "Kendalls Tau Correlation Matrix")
```
